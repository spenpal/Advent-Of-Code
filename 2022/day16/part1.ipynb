```python
# IMPORTS
import re
from pprint import pprint

import matplotlib.pyplot as plt
import networkx as nx
```

```python
# FUNCTIONS


```

```python
# MAIN
G = nx.Graph()
flows = {}

with open("input.txt") as f:
    for line in f:
        node, *neighbors = re.findall(r"[A-Z]{2}", line)
        flow = int(re.findall(r"\d+", line)[0])
        flows[node] = flow
        G.add_node(node, flow=flow)
        G.add_edges_from([(node, neighbor) for neighbor in neighbors])
        
pos = nx.spring_layout(G, k=.15, iterations=420)
plt.figure(figsize=(20,10))
nx.draw(G, pos, node_size=500, font_size=10, with_labels=True, labels=flows)
plt.show()
```
